## Pre-processing dati Alphavantage

```python
import pandas as pd
from alpha_vantage.timeseries import TimeSeries
import time
#utilizzando una free key possiamo utilizzare solo 5 query alla volta, per poter eseguire il secondo blocco attendere almeno 60 secondi
ts = TimeSeries (key='FAESZBR0KJHGLIVB', output_format='json')
ftse = ts.get_daily(symbol='^FTSE')
dax = ts.get_daily(symbol='^GDAXI')
nya= ts.get_daily(symbol='^NYA')
merv= ts.get_daily(symbol='^MERV')
nikkei = ts.get_daily(symbol='^N225')
```

```python
bsesn =ts.get_daily(symbol='^BSESN')
axjo =ts.get_daily(symbol='^AXJO')
ibex =ts.get_daily(symbol='^IBEX')
gsptse =ts.get_daily(symbol='^GSPTSE')
cac40=ts.get_daily(symbol='^FCHI')
```

```python
def convert_json(d):
    # converte la risposta in datetimerecords che possono essere
    # parsed da Pandas
    for dt, prec in d[0].items():
        r = { 'datetime': dt}
        r.update(prec)
        yield r


uk = pd.DataFrame(convert_json(ftse))
uk['datetime'] = pd.to_datetime(uk['datetime'])
#aggiungiamo due colonne utili per poter riconoscere l'indice
uk['Symbol']= '^FTSE'
uk['Country']='United Kingdom'

germany = pd.DataFrame(convert_json(dax))
germany['datetime'] = pd.to_datetime(germany['datetime'])
germany['Symbol']= '^GDAXI'
germany['Country']='Germany'

us = pd.DataFrame(convert_json(nya))
us['datetime'] = pd.to_datetime(us['datetime'])
us['Symbol']= '^NYA'
us['Country']='US'

argentina = pd.DataFrame(convert_json(merv))
argentina['datetime'] = pd.to_datetime(argentina['datetime'])
argentina['Symbol']= '^MERV'
argentina['Country']='Argentina'

japan = pd.DataFrame(convert_json(nikkei))
japan['datetime'] = pd.to_datetime(japan['datetime'])
japan['Symbol']= '^N225'
japan['Country']='Japan'

india = pd.DataFrame(convert_json(bsesn))
india['datetime'] = pd.to_datetime(india['datetime'])
india['Symbol']= '^BSESN'
india['Country']='India'

australia = pd.DataFrame(convert_json(axjo))
australia['datetime'] = pd.to_datetime(australia['datetime'])
australia['Symbol']= '^AXJO'
australia['Country']='Australia'

spain = pd.DataFrame(convert_json(ibex))
spain['datetime'] = pd.to_datetime(spain['datetime'])
spain['Symbol']= '^IBEX'
spain['Country']='Spain'

canada = pd.DataFrame(convert_json(gsptse))
canada['datetime'] = pd.to_datetime(canada['datetime'])
canada['Symbol']= '^GSPTSE'
canada['Country']='Canada'

france = pd.DataFrame(convert_json(cac40))
france['datetime'] = pd.to_datetime(france['datetime'])
france['Symbol']= '^FCHI'
france['Country']='France'

#inseriamo le serie storiche in mongo
from pymongo import MongoClient
import pymongo
client=MongoClient()
db=client.dataman
alpha = db.alpha
uk_ = uk.to_dict(orient = 'records')
result = db.alpha.insert_many(uk_ )
us_ = us.to_dict(orient = 'records')
result = db.alpha.insert_many(us_ )
japan_ = japan.to_dict(orient = 'records')
result = db.alpha.insert_many(japan_ )
germany_ = germany.to_dict(orient = 'records')
result = db.alpha.insert_many(germany_ )
argentina_ = argentina.to_dict(orient = 'records')
result = db.alpha.insert_many(argentina_ )
australia_ = australia.to_dict(orient = 'records')
result = db.alpha.insert_many(australia_ )
india_ = india.to_dict(orient = 'records')
result = db.alpha.insert_many(india )
spain_ = spain.to_dict(orient = 'records')
result = db.alpha.insert_many(spain_ )
canada_ = canada.to_dict(orient = 'records')
result = db.alpha.insert_many(canada_ )
france_ = france.to_dict(orient = 'records')
result = db.alpha.insert_many(france_ )
```

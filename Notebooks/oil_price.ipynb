```python
import datetime as dt
import pandas as pd
import pandas_datareader.data as dd
start = dt.datetime(2020,1,2)
end = dt.datetime(2020,6,2)
pd.options.display.max_rows = 999
```

```python
#prezzo al barile Brent
bz = dd.DataReader('BZ=F', 'yahoo', start, end)
brent= bz.reset_index()
brent['Date']=brent['Date'].astype(str)
brent['Date'] = pd.to_datetime(brent['Date'], format='%Y-%m-%d')
brent['Symbol']= 'BZ=F'
brent['Rate']=0
i=0
for i in range (len(brent.Close)-1):
     brent.Rate.loc[i+1]=((brent['Close'].loc[i+1]-brent['Close'].loc[i])/brent['Close'].loc[i])*100
brent= brent.drop(columns =['High','Low','Open','Volume','Adj Close'])
brent
```

```python
#prezzo al barile WTI
cl = dd.DataReader('CL=F', 'yahoo', start, end)
wti= cl.reset_index()
wti['Date']=wti['Date'].astype(str)
wti['Date'] = pd.to_datetime(wti['Date'], format='%Y-%m-%d')
wti['Symbol']= 'CL=F'
wti['Rate']=0
i=0
for i in range (len(wti.Close)-1):
     wti.Rate.loc[i+1]=((wti['Close'].loc[i+1]-wti['Close'].loc[i])/wti['Close'].loc[i])*100
wti= wti.drop(columns =['High','Low','Open','Volume','Adj Close'])
wti
```

```python
oil_price = pd.merge(wti, brent, on ='Date', suffixes = ('_WTI', '_BRENT'))
del oil_price['Symbol_WTI']
del oil_price['Symbol_BRENT']
oil_price
```

```python
oil_price['Cum_Rate_WTI'] = oil_price['Rate_WTI'].cumsum(axis = 0) 
oil_price['Cum_Rate_BRENT'] = oil_price['Rate_BRENT'].cumsum(axis = 0) 

oil_price.to_csv('/Users/enricoconte/Desktop/università/specialistica/1 Anno/1 Semestre/DataMan_DataViz/Progetto/Dati/oil_price.csv')
oil_price
```

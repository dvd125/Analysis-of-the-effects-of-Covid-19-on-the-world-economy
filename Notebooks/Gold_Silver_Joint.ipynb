```python
Silver = pd.read_csv('/Users/enricoconte/Desktop/università/specialistica/1 Anno/1 Semestre/DataMan_DataViz/Progetto/Dati/Silver.csv')
del Silver['Unnamed: 0']
Silver
```

```python
# Normalizzazione e calcolo rendimento cumulato

Silver['daily_rend'] = 0
i=0
for i in range (len(Silver.Close)-1):
     Silver.daily_rend.loc[i+1]=(((Silver['Close'].loc[i+1]*100)/17.966000)-100)
        

#del Silver['Confirmed']
Silver
```

```python
# JOIN Datasets

gold_silver = pd.merge(Gold, Silver, on= 'Date', how= 'left', suffixes= ('_Gold', '_Silver'))
gold_silver
#gold_silver.isnull().values.any()
#del gold_silver['Rate_Gold']
#del gold_silver['Rate_Silver']
#del gold_silver['Symbol_Gold']
#del gold_silver['Symbol_Silver']
gold_silver

```

```python
# Variazione giornaliera cumulativa

gold_silver['Daily_confirmed']=0
i=0
for i in range (len(gold_silver.Confirmed)-1):
    gold_silver.Daily_confirmed.loc[i]=((gold_silver['Confirmed'].loc[i+1]-gold_silver['Confirmed'].loc[i]))
gold_silver
```

```python
gold_silver['Rate_Confirmed'] = 0
i = 0
for i in range (len(gold_silver.Confirmed)-1):
     gold_silver.Rate_Confirmed.loc[i+1]=((gold_silver['Confirmed'].loc[i+1]-gold_silver['Confirmed'].loc[i])/gold_silver['Confirmed'].loc[i])*100
        
gold_silver
```

```python
gold_silver2 = gold_silver[gold_silver['Date'] > '2020-01-23']
gold_silver2.reset_index()
gold_silver2
```
